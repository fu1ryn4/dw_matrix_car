```python

(
    df
    .groupby('param_marka-pojazdu')['price_value']
    .agg([np.mean,np.median,np.size])
    .sort_values(by='mean',ascending=False)
    .head(50)
).plot(kind='bar',figsize=(15,5))
```

```python

(
    df
    .groupby('param_marka-pojazdu')['price_value']
    .agg([np.mean,np.median,np.size])
    .sort_values(by='mean',ascending=False)
    .head(50)
).plot(kind='bar',figsize=(15,5),subplots=True)
```

```python

(
    df
    .groupby('param_marka-pojazdu')['price_value']
    .agg([np.mean,np.median,np.size])
    .sort_values(by='size',ascending=False)
    .head(50)
).plot(kind='bar',figsize=(15,5),subplots=True)
```

```python
def group_and_barplot(feat_groupby, feat_agg, agg_funcs=[np.mean, np.median, np.size], feat_sort='mean', top=50, subplots=True):
    return (
        df
        .groupby(feat_groupby)[feat_agg]
        .agg(agg_funcs)
        .sort_values(by=feat_sort,ascending=False)
        .head(top)
    ).plot(kind='bar',figsize=(15,5),subplots=True)

```

```python
group_and_barplot('param_marka-pojazdu','price_value')
```

```python
group_and_barplot('param_country-of-origin','price_value')
```

```python
group_and_barplot('param_kraj-pochodzenia','price_value')
```

```python
group_and_barplot('param_kraj-pochodzenia','price_value',feat_sort='size')
```

```python
group_and_barplot('param_kolor','price_value',feat_sort='mean')
```
